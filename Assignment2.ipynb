### 1. Create the below pattern using nested for loop in Python.


```python
for i in range(1,10):
    for j in range(1,6):
        if(i>=1 and i<=5):
            if(j<=2*i-i):
                print("*",end="")
            else:
                print(" ",end="")
        elif(i>5 and i<=9):
            if(j<=10-i):
                print("*",end="")
            else:
                print(" ",end="")
    print()
```

### 2. Write a Python program to reverse a word after accepting the input from the user.

```python
string = input("Enter a String - ")
print("reverse string - " ,string[::-1])
```
